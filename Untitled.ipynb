```python
import glob
```

```python
import matplotlib.pyplot as plt
```

```python
three_files = glob.glob('slurm-647516*.out')
four_files = glob.glob('slurm-64752*.out')
five_files = glob.glob('slurm-64753*.out')[:32]
six_files = glob.glob('slurm-64753*.out')[32:] + glob.glob('slurm-64754*.out')
seven_files = glob.glob('slurm-64756*.out') + glob.glob('slurm-64757*.out')
```

```python
three_energies = []
for file in three_files:
    with open(file) as f:
        energy = float(f.readlines()[-1])
        three_energies.append(energy)

four_energies = []
for file in four_files:
    with open(file) as f:
        energy = float(f.readlines()[-1])
        four_energies.append(energy)   
        
five_energies = []
for file in five_files:
    with open(file) as f:
        energy = float(f.readlines()[-1])
        five_energies.append(energy)
        
six_energies = []
for file in six_files:
    with open(file) as f:
        energy = float(f.readlines()[-1])
        six_energies.append(energy)
        
seven_energies = []
for file in seven_files:
    with open(file) as f:
        energy = float(f.readlines()[-1])
        seven_energies.append(energy)
```

```python
plt.scatter([i for i in range(len(three_energies))], three_energies)
plt.show()
```
