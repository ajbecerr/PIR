```python
import glob
```

```python
import matplotlib.pyplot as plt
```

```python
all_files = glob.glob('*.out')
```

```python
all_energies = []
for file in all_files:
    with open(file) as f:
        energy = float(f.readlines()[-1])
        all_energies.append(energy)
```

```python
# from mpl_toolkits import mplot3d

# fig = plt.figure()
# ax = plt.axes(projection='3d')
for j in range(6):
    plt.plot([i for i in range(8)], [all_energies[8*j+i] for i in range(8)])
plt.show()
```
