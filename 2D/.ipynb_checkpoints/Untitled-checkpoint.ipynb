```python
import glob
```

```python
import matplotlib.pyplot as plt
```

```python
three_files = glob.glob('slurm-667*.out')
```

```python
three_energies = []
for file in three_files:
    with open(file) as f:
        energy = float(f.readlines()[-1])
        three_energies.append(energy)
```

```python
from mpl_toolkits import mplot3d

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.scatter([[i for j in range(8)] for i in range(8)], [[j for j in range(8)] for i in range(8)], [three_energies[8*i:8*(i+1)] for i in range(8)])
plt.show()
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D([[i for j in range(8)] for i in range(8)], [[j for j in range(8)] for i in range(8)], [three_energies[8*i:8*(i+1)] for i in range(8)])
ax.view_init(-140, 60) 
plt.show()
```
